## Load the data layer


```python
from pathlib import Path

import pandas as pd

DATA = Path.cwd().parent / "data"

# The data layer is the only permitted input. A notebook that hard-codes a
# benchmark score, a price, or an energy figure is a defect: the value would
# then exist with no source_id behind it.
models = pd.read_csv(DATA / "models.csv")
sources = pd.read_csv(DATA / "sources.csv")
print(f"{len(models)} model row(s), {len(sources)} source row(s)")
```

## Partition by evaluation conditions


```python
benchmarks = pd.read_csv(DATA / "benchmarks.csv")

# Conditions are part of the identity of a result, not metadata about it.
CONDITIONS = ["harness", "sampling_policy", "tool_permissions"]

ranked = benchmarks[~benchmarks[CONDITIONS].isin(["unstated"]).any(axis=1)]
excluded = len(benchmarks) - len(ranked)
print(f"{excluded} result(s) excluded from ranked views because conditions are unstated")

for key, group in ranked.groupby(["benchmark_id"] + CONDITIONS):
    print(key, len(group))
```
